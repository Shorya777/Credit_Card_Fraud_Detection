```python
# fig, axs = plt.subplots(1,2,figsize =(15,7))
plt.figure(figsize = (15,7))
# plt.subplot(121)
sns.histplot(data = df,x ='Amount',bins =30, color ='g')
plt.xlabel('Amount')
plt.ylabel('Class')
plt.show()
```

```python
class_0_data = df[df['Class'] == 0]
class_1_data = df[df['Class'] == 1]
plt.scatter(class_0_data['Time'], class_0_data['Amount'], color = 'blue',label = '0', alpha = 0.2)
plt.scatter(class_1_data['Time'], class_1_data['Amount'], color = 'red', label = '1', alpha = 0.2)
plt.xlabel('Amount')
plt.ylabel('Time')
plt.legend(title = 'Class', loc = 'upper right')
plt.figure(figsize = (15,7))
plt.show()
```

```python
max_amount_by_class = df.groupby('Class')['Amount'].max()
print("max amount for class 0", max_amount_by_class[0])
print("max amount for class 1", max_amount_by_class[1])
```
